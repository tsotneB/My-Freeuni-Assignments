```python
# From the previous line, we see that null values only appear in numeric columns, so it will be enough to just fill these 
# with 0-s.
# I think filling these spaces with zeroes is the most right and logical thing to do, because the best explanation for these
# null values is that maybe no one has voted for that particular political party (we see that null values only appear
# in the columns where we're supposed to see how many votes parties got in the district)
def handleNullValues(dataframe):
    for col in dataframe.columns:
        if dataframe[col].isnull().sum()>0:
            dataframe[col].fillna(0,inplace=True)
```

# For the next part, I'm going to divide our 2021 elections' data into distinct tables for proportional, majoritarian and mayor types, as it feels to be more logical and proper thing to look for anomalies in each kind of elections separately. 

```python
electionsp = electionsProp.drop(columns=['election_type'],inplace=False)
electionsp.drop(columns='district_alt_id',inplace=True)
```

```python
# using Central Limit Theorem, by subtracting mean and dividing by standart deviation (square root from variance), 
# we may expect our data's distribution to be similar to normal (gaussian distribution)
electionsp = (electionsp - electionsp.mean())/(electionsp.std())
```

```python
handleNullValues(electionsp)
```

```python
#I'm going to try two approaches for our anomaly detection.
# first of all, I'm going to construct gaussian distribution for every single column in our data
# I also added column for probabilities, for which the initial values are 1. (they will be computed later)
everyColumnGaussianProp = np.exp(np.square(electionsp)/(-2))/(np.sqrt(2*np.pi))
everyColumnGaussianProp['Probabilities']=1
```

```python
# Here we compute the probabilities of each row. Because we consider these columns independent, there joint pmf is 
# product of marginal pmfs. 
for i in everyColumnGaussianProp:
    everyColumnGaussianProp['Probabilities'] *= everyColumnGaussianProp[i]
```

```python
# In this part, we are in business with really small values of probabilities, so our treshold should be really small if we 
# want our model to detect strange cases
# I've chosen 1e-100 as treshold, which is pretty small number and close to zero
anomalyPerc = 1e-100
everyColumnGaussianProp[everyColumnGaussianProp['Probabilities'] < anomalyPerc]
```

```python
# We've seen that for 1e-100, our model considered 21 cases out of 3736 to be anomalistic
# Here, we see that for 1e-50, which is also very small number (But big compared to 1e-100), number of anomalies spotted
# are ten times bigger
anomalyPerc = 1e-50
everyColumnGaussianProp[everyColumnGaussianProp['Probabilities'] < anomalyPerc]
```

# So far, I've constructed gaussian distribution with all the columns of our data and tried to find anomalies in proportional elections. for different values of tresholds, algorithm has found either 21 (treshold set at 1e-100) or 119 (1e-50) anomalies.
